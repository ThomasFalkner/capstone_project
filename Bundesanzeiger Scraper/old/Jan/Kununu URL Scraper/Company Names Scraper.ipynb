```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import os
from tabulate import tabulate
import json
from win32process import CREATE_NO_WINDOW
import time
import numpy as np
from selenium.common import exceptions
import json


def getDriver():
    
    chrome_options = Options()
    chrome_options.headless = True
    chrome_options.add_argument("--headless")
    chrome_options.add_argument("--hidecommandpromptwindow")
    chrome_driver = os.getcwd() +"\\chromedriver.exe"
    args = ["hide_console", ]
    return webdriver.Chrome(options=chrome_options, executable_path=chrome_driver, service_args=args)

    
    
def scrape_names():
    with open('IT/urls_merged.json','r') as file: 
        url_list = json.load(file)
    name_list = []
    driver = getDriver()
    for url in url_list:
        driver.get(url)
        try:
            name_field = driver.find_element_by_xpath("/html/body/main/div/div[1]/div[1]/h1/span[1][@itemprop='name']")
        except exceptions.NoSuchElementException as e: 
            continue;
        name = name_field.text 
        print(name)
        #name = getNameFromUrl(url); 
        name_list.append(name)
    
    return name_list
        
names = scrape_names()


with open('names.json','w') as f:
   json.dump(list(names),f)
```
